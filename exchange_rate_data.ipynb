```python
import pandas as pd
import numpy as np
```

```python
canada_df = pd.read_csv("data/exchange-rates/canada-exchange-rate.csv", skiprows=3)
france_df = pd.read_csv("data/exchange-rates/france-exchange-rate.csv", skiprows=3)
germany_df = pd.read_csv("data/exchange-rates/germany-exchange-rate.csv", skiprows=3)
italy_df = pd.read_csv("data/exchange-rates/italy-exchange-rate.csv", skiprows=3)
uk_df = pd.read_csv("data/exchange-rates/uk-exchange-rate.csv", skiprows=3)
usa_df = pd.read_csv("data/exchange-rates/usa-exchange-rate.csv", skiprows=3)
japan_df = pd.read_csv("data/exchange-rates/japan-exchange-rate.csv", skiprows=3)
```

```python
canada_df
```

```python
canada_df = canada_df[["TIME_PERIOD:Period", "OBS_VALUE:Value", "Unit"]]
italy_df = italy_df[["TIME_PERIOD:Period", "OBS_VALUE:Value", "Unit"]]
france_df = france_df[["TIME_PERIOD:Period", "OBS_VALUE:Value", "Unit"]]
germany_df = germany_df[["TIME_PERIOD:Period", "OBS_VALUE:Value", "Unit"]]
uk_df = uk_df[["TIME_PERIOD:Period", "OBS_VALUE:Value", "Unit"]]
usa_df = usa_df[["TIME_PERIOD:Period", "OBS_VALUE:Value", "Unit"]]
japan_df = japan_df[["TIME_PERIOD:Period", "OBS_VALUE:Value", "Unit"]]
```

```python
df_list = [canada_df, italy_df, france_df, germany_df, uk_df, usa_df, japan_df]
```

```python
for country in df_list:
    country['TIME_PERIOD:Period'] = pd.to_datetime(country['TIME_PERIOD:Period'], format= "%Y-%m-%d")
    country['TIME_PERIOD:Period'] = country['TIME_PERIOD:Period'].dt.strftime("%Y-%m")
    country.rename(columns={'TIME_PERIOD:Period': "TIME_PERIOD", "OBS_VALUE:Value": "exchange_rate"}, inplace=True)
```
