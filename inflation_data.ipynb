```python
import pandas as pd
```

```python
data = pd.read_csv("./data/inflation/OECD,DF_DP_LIVE,+IND+BRA+CHN+DEU+FRA+JPN+ITA+CAN+USA+GBR.CPI...M.csv")
```

```python
data = data[['TIME_PERIOD', 'Country', 'INDICATOR', 'Subject', 'OBS_VALUE', 'MEASURE']]
```

```python
data.sort_values("TIME_PERIOD", inplace=True)
```

```python
data.info()
```

```python
data.tail(20)
```

```python
data["MOM"] = data.apply(lambda x: x['OBS_VALUE'] if x['MEASURE'] == "AGRWTH" else None, axis = 1)
```

```python
data.tail(20)
```

```python
data_temp = data[['TIME_PERIOD', 'Country', 'OBS_VALUE', 'MEASURE']]
```

```python
data_temp
```

```python
data_temp = data_temp.drop(data_temp[data_temp["MEASURE"] == "IDX2015"].index , axis = 0)
```

```python
data_temp
```

```python
data = data.drop(data[data["MEASURE"] == "AGRWTH"].index , axis = 0)
data
```

```python
data = pd.merge(data, data_temp, on= ["TIME_PERIOD", "Country"], how = "inner")
```

```python
data
```

```python
# Pivot the DataFrame to convert the 'Subject' values into columns
pivoted_x = data.pivot_table(index=['TIME_PERIOD', 'Country'], columns='Subject', values='OBS_VALUE_x', aggfunc = 'mean')
```

```python
pivoted_x = pivoted_x.reset_index()
pivoted_x
```

```python
# Pivot the DataFrame to convert the 'Subject' values into columns
pivoted_y = data.pivot_table(index=['TIME_PERIOD', 'Country'], columns='Subject', values='OBS_VALUE_y', aggfunc = 'mean')
```

```python
pivoted_y = pivoted_y.reset_index()
```

```python
pivoted_y
```

```python
pivoted_x.columns = [f"{col}_IDX2015" for col in pivoted_x.columns]
pivoted_y.columns = [f"{col}_MoM" for col in pivoted_y.columns]
```

```python
final_data = pd.concat([pivoted_x, pivoted_y], axis=1).reset_index()
```

```python
final_data
```

```python
final_data.rename(columns={"TIME_PERIOD_IDX2015" : "date", "Country_IDX2015": "country", "Total less food, less energy_IDX2015": "core_IDX2015",
                            "Total less food, less energy_MoM": "core_MoM" }, inplace=True)
```

```python
final_data['index']
```

```python
final_data = final_data.set_index("date")
final_data = final_data.drop("index", axis = 1)
final_data = final_data.reset_index()
```

```python
final_data
```

```python
final_data.drop(["TIME_PERIOD_MoM", "Country_MoM"], axis=1, inplace=True)
```
