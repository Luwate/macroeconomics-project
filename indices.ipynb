We convert the stock indices data to percentage growth as the indices are in different currencies and have different magnitudes thus, MoM increments are better to use

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
import yfinance as yf

# Define the stock ticker
canada = "^GSPTSE"
italy = "FTSEMIB.MI"
germany = "^GDAXI"
france = "^FCHI"
uk = "^FTSE"
usa = "^GSPC"
japan = '^N225'

index_list = [canada, italy, germany, france, uk, usa, japan]
```

```python
data = yf.download(canada, start='1940-01-01', end='2024-09-01', interval='1d')

    # View the data
print(data.head())

    # Save to a CSV file
data.to_csv(f'./data/indices/{canada}_historical_data.csv')
```

```python
for index in index_list:
    data = yf.download(index, start='1940-01-01', end='2024-09-01', interval='1d')

    # View the data
    print(data.head())

    # Save to a CSV file
    data.to_csv(f'./data/indices/{index}_historical_data.csv')
```

```python
canada_df = pd.read_csv("data/indices/^GSPTSE_historical_data.csv")
italy_df = pd.read_csv("data/indices/FTSEMIB.MI_historical_data.csv")
france_df = pd.read_csv("data/indices/^FCHI_historical_data.csv")
germany_df = pd.read_csv("data/indices/^GDAXI_historical_data.csv")
uk_df = pd.read_csv("data/indices/^FTSE_historical_data.csv")
usa_df = pd.read_csv("data/indices/^GSPC_historical_data.csv")
japan_df = pd.read_csv("data/indices/^N225_historical_data.csv")
```

```python
df_list = [canada_df, italy_df, france_df, germany_df, uk_df, usa_df, japan_df]
```

```python
for df in df_list:
    df["Date"] = pd.to_datetime(df["Date"], format="%Y-%m-%d")
```

```python
country_names = ["Canada", "Italy", "France", "Germany", "United Kingdom", "United States", "Japan"]
```

```python
scaler = StandardScaler()
scaled_dfs = []

for df, country in zip(df_list, country_names):
    df = df.copy()  # Avoid modifying the original DataFrame
    df["country"] = country  # Add country column
    numeric_cols = ["Open", "High", "Low", "Close", "Adj Close", "Volume"]  # Select numeric columns
    
    # Apply standardization (z-score)
    df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
    
    scaled_dfs.append(df)
```

```python
merged_df = pd.concat(scaled_dfs, ignore_index=True)
```

```python
merged_df
```

```python
merged_df = merged_df[merged_df["Date"].isin(df_min["max"])]
```

```python
merged_df["Date"] = pd.to_datetime(merged_df["Date"])

# Convert to Year-Month format
merged_df["Date"] = merged_df["Date"].dt.to_period("M")  # Keeps it as Period format (YYYY-MM)

# If you need it as a string (optional)
merged_df["Date"] = merged_df["Date"].dt.strftime("%Y-%m")
```

```python
merged_df.to_csv("data/indices/standardized_stock_indices.csv", index=False)
```
